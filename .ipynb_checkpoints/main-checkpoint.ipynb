```python
#Importing the Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Models used for Training
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Score Evaluator based on Output
from sklearn.metrics import classification_report, accuracy_score
```

```python
#Reading the data with Pandas library
data_set = pd.read_csv("breast_cancer_detection.csv")

#Seperating the Data Sets into Dependent -> (y) and Independent -> (X) Variables.
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values

#Displaying the first 10 Rows of the Data Set. 
data_set.head(10)
```

```python
#Calculating the Rows and Columns:
data_set.shape
```

```python
#To get the description and idea of Table Values
data_set.describe()
```

```python
#Checking if there is any Empty Cells,So that Data Pre-processing Can be done..
data_set.isna().sum()
#Thankfully there is no Empty cells.
```

```python
#Check the Correlations of the DataSet:
#The number Closer to 1, more the  Row is Correlated with Column.
#Negative number indicates they're negatively Correlated.
X_corr = data_set.iloc[:,:11] # Storing and displaying only Mean values and omitting worst Values.
X_corr.corr()
```

```python
#Visualize the Correlation using Seaborn.
#The Colors indicate how closely or farely theyre related with each other.

plt.figure(figsize=(10,10)) #Increasing the size of fields for better view
sns.heatmap(X_corr.corr(), annot = True, fmt = ".0%")
```

```python
#Splitting the Data for Training, Testing such that we're able to sufficiently let the AI understand the datS 
x_train, x_test, y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)
```

```python
#Scaling all the values in table to a Standard -3 to 3 range.
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#Creating different Models Function and Predicting Who's Accuracy is better. 

def Models(x_train,y_train):
    log_reg = LogisticRegression(random_state = 15)
    log_reg.fit(x_train,y_train)
    
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state = 15)
    tree.fit(x_train,y_train)
    
    Forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 15)
    Forest.fit(x_train,y_train)
    print("Training Data")
    print()
    print("Logistic Regression's Accuracy is : ",log_reg.score(x_train,y_train))
    print("Decision Tree Classifier Accuracy is : ",tree.score(x_train,y_train))
    print("Random Forest Classifier's Accuracy is : ",Forest.score(x_train,y_train))
    print()
    return log_reg, tree, Forest
```

```python
models = Models(x_train,y_train)
model_name = ["Logistic Regression","Decision Tree Classifier","Random Forest Classifier"]
print("Test Data")
print()
for i in range(len(models)):
    print(model_name[i], "Accuracy is : ", models[i].score(x_test,y_test))
```

```python
# From the Above Cell Output Even though the Decision Tree performed well in Training Sprints,
# In Test Sprint Logistic Regression performed slightly better than Decision Tree

# So, Implementing Logistic Regression is the Best Case.. 
```

```python
# y_pred is Our Model's Prediction and y_test is the Actual Output
y_pred = models[0].predict(x_test)

#Combining y_pred, y_test so we could check if they match. 
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test),1)) , 1))
```

```python
print(classification_report(y_test, y_pred))
print()
print("Accuracy Score is : ",accuracy_score(y_test, y_pred))
```

```python
correct_prediction = accuracy_score(y_test,y_pred)
error_prediction = 1-correct_prediction

pred_arr = [correct_prediction,error_prediction]
label = ["Accurate", "Error"]
colors = ["c","r"]

#Visualizing Error and Correct Ratio in Pie Chart.
plt.title("Error Representation")
plt.pie(pred_arr,colors = colors, labels = label,autopct = "%.2f%%",
        startangle = 90, 
        wedgeprops = {"edgecolor" : "black",
                      'linewidth': 1,
                      'antialiased': True})
```
